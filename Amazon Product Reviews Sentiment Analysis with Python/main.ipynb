```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import warnings
warnings.filterwarnings("ignore")
import nltk
nltk.download("all")
```

```python
#Task 1
#Now let’s start this task of Amazon review sentiment analysis by importing the necessary Python libraries and the dataset:
def load_the_dataset():
    df = pd.read_csv("Reviews.csv")
    return df

#Task 2
# As this dataset is very large, it contains some missing values, so let’s remove all the rows containing the missing values:
def pre_processing():
    df = load_the_dataset()
    df_1 = df.dropna(axis=0,inplace=False)
    return df_1


# Task 3
# The Score column of this dataset contains the ratings that customers have given to the product based on their experience with the product.
# So let’s take a look at the rating breakdown to see how most customers rate the products they buy from Amazon:
def distribution_of_ratings():
    df_1 = pre_processing()
    values = df_1["Score"].value_counts()
    colours = ["red","green","blue","yellow","orange"]
    plt.pie(x=values,labels=values.index,colors=colours,autopct="%1.1f%%",textprops={"color":"white","weight":"bold"})
    return values



# Task 4
# Now, I’m going to add three more columns to this dataset as Positive, Negative, and Neutral by calculating the sentiment scores of the customer reviews mentioned in the Text column of the dataset:
def add_new_columns():
    df_1 = pre_processing()
    df_2 = df_1.copy()
    sia = SentimentIntensityAnalyzer()
    positive = []
    neutral = []
    negative = []
    for i in list(df_2.index):
        pos = sia.polarity_scores(df_2.loc[i,"Text"])
        positive.append(pos["pos"])
        neutral.append(pos["neu"])
        negative.append(pos["neg"])
    df_2["Positive"] = positive
    df_2["Negative"] = negative
    df_2["Neutral"] = neutral
    return df_2
```

```python
df = add_new_columns()
df
```
